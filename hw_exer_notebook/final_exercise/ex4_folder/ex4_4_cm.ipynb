```python
import re
import operator

f = open("/Users/yuchangmin/Documents/Medical Genetics/humanDNAsample.dna", 'r')
f1 = f.readlines()
f1_new = []
for line in f1:
    line = line.strip()
    f1_new.append(line)
fin_file = []
i = 8
while i < len(f1_new):
    fin_file.append(f1_new[i])
    i += 1
f_str = "".join(fin_file)
final = []
for i in range(len(f_str)-3):
    final.append(f_str[i:i+4])
fin_dic = {}
for line in final:
    if line not in fin_dic.keys():
        fin_dic[line] = 1
    else:
        fin_dic[line] += 1
finfin = sorted(fin_dic.items(), key = operator.itemgetter(1), reverse = True)
for li in finfin:
    print(li[0],":",li[1])

```
