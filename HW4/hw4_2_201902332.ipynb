```python
import re
f = open("/Users/yuchangmin/Documents/Medical Genetics/HW4/sample.bed", 'r')
f1 = f.readlines()
f1_new = []
for line in f1:
    once = []
    once = line.split('\t')
    f1_new.append(once) ## f1 line tab(\t)   once  .   f1_new append     
f1_new.sort(key=lambda x:x[4])
i = 0
j = 0
k = 0
f_dic = {}
str_dic = {}
chr_dic = {}
chr_list = []
str_list = []
while i < len(f1):
    li = f1_new[i][4].strip()
    uniq = f1_new[i][5].strip()
    if li not in f_dic.keys():
        f_dic[li] = uniq
    elif li in f_dic.keys():
        f_dic[li] += uniq
    i += 1 ##  [i][4]  , [5]  sequence  \n (strip).   if li f_dic  uniq value  ,    value  
while j < len(f1):
    strand = f1_new[j][3].strip()
    li = f1_new[j][4].strip()
    str_dic[li] = strand
    j += 1 
while k < len(f1):
    chrome = f1_new[k][0].strip()
    li = f1_new[k][4].strip()
    chr_dic[li] = chrome
    k += 1 
final_list = []
for key in f_dic.keys():
    app = []
    app.append(chr_dic[key])
    app.append(key)
    app.append(str_dic[key])
    app.append(f_dic[key])
    final_list.append(app)
for lin in final_list:
    kk = "\t".join(lin)
    print(kk)
```
