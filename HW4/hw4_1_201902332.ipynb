```python
import re

f = open("/Users/yuchangmin/Documents/Medical Genetics/HW4/SP1.fq", 'r')
f1 = f.readlines()
f1_new = []
f2 = []
fnew = []
f2new = []
i=0     
j=1
k=0
while i < len(f1):
    m = re.search('U.*\n', f1[i]).group()
    f1_new.append(m)
    i += 4 ## UMI로 시작하는 문자열을 뽑아내 m에 저장하고 그것을 f1_new 리스트에 append / +=4하는 이유는 0 4 8 12 ...번째 배열에 umi라인이 존재하기 때문
for line in f1_new:
    line = line.strip()
    fnew.append(line) ## \n을 지워서 fnew 리스트에 append
while j < len(f1):
    li = f1[j]
    f2.append(li)
    j += 4 ## 윗윗 while문과 동일한 방식으로 sequence f2에 append
for line2 in f2:
    line2 = line2.strip()
    f2new.append(line2) ## \n을 지워서 f2new 리스트에 append
f_newdic = {}
q = 0
for k in fnew:
    idf = k
    if idf not in f_newdic.keys():
        f_newdic[idf] = f2new[q]
    elif idf in f_newdic.keys():
        f_newdic[idf] += f2new[q]
    q += 1
k_list = list(f_newdic.keys())
sort_list = sorted(k_list)
for sort_key in sort_list:
    print(sort_key,":",f_newdic[sort_key])
```
